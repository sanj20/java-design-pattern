```
import numpy as np
import matplotlib.pyplot as plt
import seaborn 
%matplotlib inline
```

```
x = np.linspace(-10,10,num=1000)
plt.figure(figsize=(12,8))
plt.plot(x,1/(1+np.exp(-x)));## sigmoid function
plt.title("sigmoid function")
```
